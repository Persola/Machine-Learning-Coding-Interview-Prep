```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def make_data():
    m = np.random.uniform(-3, 3)
    return np.random.normal(m, 2, 10000)
```

## Cumulative distribution function

The *cumulative distribution function (CDF)* of a random variable $y$ is a function $\Phi$ from $\mathbf{R}$ to $[0, 1]$. The number $\Phi(x)$ represents the probability that you will draw a number less than or equal to $x$, that is,
$$ \Phi(x) = \mathbf{P(y \le x)}.$$
For example, if $y$ is distrubuted uniformly on the interval $[5, 6]$, $\Phi$ is zero for $x < 5$, goes linearly from zero to one over the interval $[5, 6]$, and is one for $x > 6$.

The *empirical CDF* of a set $Y$ of $N$ random numbers is defined similarly: $\Phi(x)$ is the probability that a randomly selected element of $Y$ is less than or equal to $x$. That is, $\Phi(x)$ the number of elements of $Y$ less than or equal to $x$, divided by $N$.

### Question 1

Write a function that takes an array of numbers as input and returns their CDF. How will you represent the CDF -- is it also a function, an array, or what? Explain your choice.

```python
Y = make_data()
```

### Answer 1.1

One way is to make a function that depends on $Y$ and $x$:

```python
def cdf(Y, x):
    return np.sum(Y <= x) / len(Y)
```

```python
xx = np.linspace(-10, 10, 50)
C = [cdf(Y, x) for x in xx]
```

```python
plt.plot(xx, C)
```

### Answer 1.2

Or, you can return an array, along with sample points:

```python
def cdf(Y, k):
    # CDF of Y at k points
    xx = np.linspace(min(Y), max(Y), k, endpoint=True)
    sorted_Y = np.sort(Y)
    ii = np.searchsorted(sorted_Y, xx)
    return xx, ii/len(Y)
```

```python
xx, C = cdf(Y, 50)
```

```python
plt.plot(xx, C)
```

### Answer 1.3

One more way: use $Y$ as the sample points, allowing uneven sample points:

```python
def cdf(Y):
    sorted_Y = np.sort(Y)
    return sorted_Y, np.linspace(0, 1, len(Y))
```

```python
xx, C = cdf(Y)
```

```python
plt.plot(xx, C, ".")
```

### Question 3

Suppose $Y$ and $Z$ are sets of measured wait times in seconds between transactions on two different database servers, and a co-worker asks you if the distribution of load on the two servers is the same or not. How would you tackle this question -- how would you make it precise, and how would you provide a quantitative answer?

Use the provided data to illustrate your method.

```python
Y = make_data()
Z = make_data()
```

### Answer 3

One precise version of the question is, "Do $Y$ and $Z$ have the same distribution?" -- $Y$ and $Z$ are both sets of random numbers, and if the distributions are different, you can say the two servers have different loads.

We can use the CDFs of $Y$ and $Z$ to answer the question quantitatively. In practice, you can start by plotting them both; if the CDFs look different, the distributions are probably different.

```python
xxY, CY = cdf(Y)
xxZ, CZ = cdf(Z)
```

```python
plt.plot(xxY, CY)
plt.plot(xxZ, CZ)
```

The *Kolmogorov-Smirnov (KS) test* is one way to quantify this idea. To do the KS test, find the $x$ where the CDFs of $Y$ and $Z$ are the furthest apart; the absolute difference in CDFs at this point is the KS statistic. The higher the statistic, the less likely $Y$ and $Z$ are to have come from the same distribution.

There are other variations on this idea that also can work; the KS test isn't the only good option.

```python
def ks(Y, Z):
    xmin = min(min(Y), min(Z))
    xmax = max(max(Y), max(Z))
    K = max(len(Y), len(Z))
    xx = np.linspace(xmin, xmax, K)
    sorted_Y = np.sort(Y)
    CY = np.searchsorted(sorted_Y, xx) / K
    sorted_Z = np.sort(Z)
    CZ = np.searchsorted(sorted_Z, xx) / K
    return np.max(np.abs(CY - CZ))
```

```python
ks(Y, Z)
```

There is a formula that lets you use the KS test formally, rejecting the hypothesis that $Y$ and $Z$ come from the same different distribution at a given level, etc., but it isn't reasonable to remember how to do that on an interview. It's OK just to get the main idea.
