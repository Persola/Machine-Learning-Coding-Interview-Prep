```python
N = 1000
w0 = np.array([1, 2])
X = np.empty((N, 2))
X[:, 1] = 1 # for b
X[:, 0] = np.random.normal(0, 1, N)
y0 = X @ w0 #w0 does not have to be transposed
y = y0 + np.random.normal(0, 0.1, N)
```

```python
w0
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X[:, 0], y0, ".", label="no noise") # no noise
ax.plot(X[:, 0], y, ".", alpha=0.25, label="noise") # noise
ax.legend(loc=0)
```

```python
def mse(x, w, y):
    print((x[:3] @ w - y[:3])**2)
    return np.mean((x @ w - y)**2)

w_from_math = np.linalg.inv((X.T @ X)) @ X.T @ y
mse(X, w_from_math, y)
```

```python
def mse_grad(x, w, y):
    return np.array([
        np.mean(2 * (X@w - y) * X[:, 0]),
        np.mean(2 * (X@w - y) * X[:, 1])
    ])
```

```python
mse_grad(X, 13*w0, y0)
```

```python
def mse_grad_2(x, w, y):
    return 2 * x.T @ (x @ w - y) / N

mse_grad_2(X, 13*w0, y0)
```

```python
gradient_descent2(lambda w : mse(X, w, y),
                  lambda w : mse_grad_2(X, w, y),
                  0.2, np.zeros(2))
```
