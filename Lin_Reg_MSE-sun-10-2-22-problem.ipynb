```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
plt.style.use("bmh")
```

```python
N = 1000
w0 = np.array([1, 2])
X = np.empty((N, 2))
X[:, 1] = 1 # for b
X[:, 0] = np.random.normal(0, 1, N)
y0 = X @ w0 #w0 does not have to be transposed
y = y0 + np.random.normal(0, 0.1, N)
```
