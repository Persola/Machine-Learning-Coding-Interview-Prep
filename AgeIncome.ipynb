```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 1000
mu_age_house, sigma_age_house = 50, 10
age_house = np.random.normal(mu_age_house, sigma_age_house, N//2)
age_house = np.round(age_house)
age_house = np.clip(age_house, 19, 99)

mu_income_house, sigma_income_house = 90000, 10000
income_house = np.random.normal(mu_income_house, sigma_income_house, N//2)
income_house = np.round(income_house)
income_house = np.clip(income_house, 15000, 250000)

mu_age_rent, sigma_age_rent = 35, 10
age_rent = np.random.normal(mu_age_rent, sigma_age_rent, N//2)
age_rent = np.round(age_rent)
age_rent = np.clip(age_rent, 19, 99)

mu_income_rent, sigma_income_rent = 50000, 20000
income_rent = np.random.normal(mu_income_rent, sigma_income_rent, N//2)
income_rent = np.round(income_rent)
income_rent = np.clip(income_rent, 15000, 250000)
X_house = np.vstack((age_house, income_house)).T
X_rent = np.vstack((age_rent, income_rent)).T
X = np.concatenate((X_house, X_rent))
Y = np.zeros(N, dtype=np.bool_)
Y[:N//2] = 1
```
