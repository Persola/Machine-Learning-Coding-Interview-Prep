```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import betabinom
```

```python
%matplotlib inline
```

```python
N = 100000
scores_rejected = betabinom.rvs(100, 10, 7, 0, N//2)
scores_accepted = betabinom.rvs(100, 22, 8, 0, N//2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
_ = ax.hist(scores_rejected,
            bins=np.linspace(0, 100, 100, endpoint=True),
            alpha=0.4)
_ = ax.hist(scores_accepted,
            bins=np.linspace(0, 100, 100, endpoint=True),
            alpha=0.4)
```
